```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('data.csv', delimiter=",", quoting=3)

data_c = pd.DataFrame()

data_c['date'] = data['"data'].str.replace('"', '')           #from str to float
data_c[''] = data['""nome""'].str.replace('"', '')
data_c['celsius'] = data['""valore"""'].str.replace('"', '')

data_c = data_c.pivot(index='date', columns='', values='celsius')
data_c.drop(columns='prec', inplace=True)
data_c.rename(columns={'tempMax':'max', 'tempMin':'min'}, inplace=True)
data_c['max'] = pd.to_numeric(data_c['max'])
data_c['min'] = pd.to_numeric(data_c['min'])

data_c['DAT'] = data_c[['max', 'min']].mean(axis=1)      #1 = rows 0 = columns

data_c.index = pd.to_datetime(data_c.index)

#data_c.index = data_c.index.strftime('%d/%m/%Y')

data_c
```

```python
data1 = pd.read_csv('data1.csv', delimiter=",", quoting=2)

data1_c = pd.DataFrame()

data1_c[['date', '']] = data1['Data-Ora'].str.split(' ', expand=True)

data1[' Medio'] = pd.to_numeric(data1[' Medio'])

data1_c['celsius'] = data1[' Medio']

data1_c = data1_c.pivot(index='date', columns='', values='celsius')

data1_c['max'] = data1_c[:].max(axis=1)
data1_c['min'] = data1_c[:].min(axis=1)
data1_c['DAT'] = data1_c[['max', 'min']].mean(axis=1)

data1_c.index = pd.to_datetime(data1_c.index)
data1_c.sort_index(inplace=True)
#data1_c.index = data1_c.index.strftime('%d/%m/%Y')

data2_c = pd.DataFrame(data1_c.drop(data1_c.columns[:24], axis=1))

data2_c
```

```python
result = pd.concat([data_c, data2_c])       #data c + data 2c

result.to_csv('time-series.csv')
```
